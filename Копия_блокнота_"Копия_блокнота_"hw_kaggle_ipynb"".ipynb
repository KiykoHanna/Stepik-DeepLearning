# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

(более сложный вариант) Если вы будете использовать кроссвалидацию, то преобразования данных и LogisticRegression нужно соединить в один Pipeline с помощью make_pipeline, как это делалось во втором семинаре. Для one-hot-encoding'a можно испльзовать комбинацию LabelEncoder + OneHotEncoder (сначала превращаем строчки в числа, а потом числа првращаем в one-hot вектора.)

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
# 1. Обрабатываем данные
# Нормируем числовые признаки
scaler = StandardScaler()
data_r = pd.DataFrame(scaler.fit_transform(data[num_cols]), columns= num_cols)

# Категориальные признаки кодируем с помощью one-hot-encoding'а. 
enc = OneHotEncoder()
enc.fit(data[cat_cols])
a = enc.transform(data[cat_cols]).toarray()
dd = pd.DataFrame(a).join(data_r)
dd.columns = dd.columns.astype(str)

#  Разделяем выборку на train/valid 
X_train, X_test, y_train, y_test = train_test_split(dd, data[target_col], test_size= 0.2)
X_test, X_val, y_test, y_val = train_test_split(X_train, y_train, train_size=0.4)

# Используем LogisticRegressionCV
LL = LogisticRegressionCV(cv=5, scoring='roc_auc',  refit=True)
LL.fit(X_train, y_train)
y_pred = LL.predict_proba(X_test)
y_pred = y_pred[:, 1] 

print(LL.get_params(), '\n',  roc_auc_score(y_test, y_pred))

```

```python
# 1. Разделим выборку на train/valid. 
# Tестируйем catboost cо стандартными параметрами.
from catboost import CatBoostClassifier

X_train, X_test, y_train, y_test = train_test_split(data[feature_cols], data[target_col], test_size= 0.2, )
X_test, X_val, y_test, y_val = train_test_split(X_train, y_train, train_size=0.4)

cat = CatBoostClassifier()
cat.fit(X_train,y_train, cat_features=cat_cols, eval_set=(X_test, y_test),  verbose=False, plot=True)
```

```python
# Получаем предсказания и тестируем модель с помощью roc_auc
preds_class = cat.predict(X_test)
preds_proba = cat.predict_proba(X_test)

roc_auc_score(y_test, preds_class)
```

```python
# 2. Тестируем разные параметры модели
cat_param = CatBoostClassifier(iterations=2500,
                               learning_rate=0.3, )
cat_param.fit(X_train,y_train, cat_features=cat_cols, eval_set=(X_test, y_test), verbose=False, plot=True)
preds_class = cat_param.predict(X_test)
preds_proba = cat_param.predict_proba(X_test)
print(roc_auc_score(y_test, preds_class))
```
