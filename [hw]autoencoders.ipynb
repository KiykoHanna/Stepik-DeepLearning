Реализуем autoencoder. Архитектуру (conv, fully-connected, ReLu, etc) можете выбирать сами. Экспериментируйте!

```python
from copy import deepcopy


class Autoencoder(nn.Module):
    def __init__(self):
        super().__init__()

        self.flatten = nn.Flatten()

        # encoder
        self.encoder = nn.Sequential(
            nn.Linear(in_features=12288, out_features=512),
            nn.ReLU(),
            nn.Linear(in_features=512, out_features= dim_code)
        )
 
        # decoder 
        self.decoder = nn.Sequential(
            nn.Linear(in_features=dim_code, out_features=512),
            nn.ReLU(),
            nn.Linear(in_features=512, out_features=12288)
        )

    # sampler
    def sample(self, z):
        generated = self.decoder(z)
        generated = torch.sigmoid(generated)
        generated = generated.view(-1, 64, 64, 3)
        return generated

    def forward(self, x):
        x = self.flatten(x).float()
        latent_code = self.encoder(x)
        reconstruction = torch.sigmoid(self.decoder(latent_code))
        return reconstruction, latent_code
```

```python
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'

criterion = F.mse_loss
autoencoder_V = Autoencoder().to(device)

optimizer = torch.optim.Adam(autoencoder_V.parameters())
device
```

## 2.1 Архитектура модели и обучение (2 балла)

Реализуем VAE. Архитектуру (conv, fully-connected, ReLu, etc) можете выбирать сами. Рекомендуем пользоваться более сложными моделями, чем та, что была на семинаре:) Экспериментируйте!

```python
latent_dim = 4

class VAE(nn.Module):
    def __init__(self):
        super(VAE, self).__init__()

        # encoder
        self.conv1 = nn.Conv2d(in_channels=1, 
                               out_channels=64, 
                               kernel_size=4, 
                               stride=2, 
                               padding=1) 

        self.conv2 = nn.Conv2d(in_channels=64, 
                               out_channels=64*2, 
                               kernel_size=4, 
                               stride=2, 
                               padding=1) 

        self.fc_mu = nn.Linear(in_features=64*2*7*7, 
                               out_features=latent_dim)
        self.fc_logvar = nn.Linear(in_features=64*2*7*7, 
                                   out_features=latent_dim)
            
        self.activation = nn.ReLU()

        # decoder
        self.fc_d = nn.Linear(in_features=latent_dim, 
                            out_features=64*2*7*7)

        self.conv2_d = nn.ConvTranspose2d(in_channels=64*2, 
                                        out_channels=64, 
                                        kernel_size=4, 
                                        stride=2, 
                                        padding=1)
        self.conv1_d = nn.ConvTranspose2d(in_channels=64, 
                                        out_channels=1, 
                                        kernel_size=4, 
                                        stride=2, 
                                        padding=1)
        self.activation_d = nn.ReLU()
        
    def encode(self, x):
        x = self.activation(self.conv1(x))
        x = self.activation(self.conv2(x))

        x = x.view(x.shape[0], -1)

        x_mu = self.fc_mu(x)
        x_logvar = self.fc_logvar(x)
        return x_mu, x_logvar
     
    def decode(self, x):
        x = self.fc_d(x)
        x = x.view(x.size(0), 64*2, 7, 7)
        x = self.activation_d(self.conv2_d(x))
        x = torch.sigmoid(self.conv1_d(x))
        return x

    def gaussian_sampler(self, mu, x_logvar):
        if self.training:
                std = (x_logvar * 0.5).exp()  
                return torch.distributions.Normal(loc=mu, scale=std).rsample()
        else:
                return mu

    def reparameterize(self, mu, x_logvar):
        std = torch.exp(0.5 * x_logvar) # standard deviation
        eps = torch.randn_like(std) # `randn_like` as we need the same size
        sample = mu + (eps * std) # sampling as if coming from the input space
        return sample

    def get_latent_vector(self, x):
        mu, x_logvar = self.encode(x)
        z = self.reparameterize(mu, x_logvar)
        return z

    def forward(self, x):
        mu, x_logvar = self.encode(x)
        reconstruction = self.decode(self.gaussian_sampler(mu, x_logvar))
        return mu, x_logvar, reconstruction
```

Определим лосс и его компоненты для VAE:

Надеюсь, вы уже прочитали материал в towardsdatascience (или еще где-то) про VAE и знаете, что лосс у VAE состоит из двух частей: KL и log-likelihood.

Общий лосс будет выглядеть так:

$$\mathcal{L} = -D_{KL}(q_{\phi}(z|x)||p(z)) + \log p_{\theta}(x|z)$$

Формула для KL-дивергенции:

$$D_{KL} = -\frac{1}{2}\sum_{i=1}^{dimZ}(1+log(\sigma_i^2)-\mu_i^2-\sigma_i^2)$$

В качестве log-likelihood возьмем привычную нам кросс-энтропию.

```python
def KL_divergence(mu, logsigma):
    loss = -0.5 * torch.sum(1 + logsigma - mu.pow(2) - logsigma.exp())
    return loss

def log_likelihood(x, reconstruction):
    loss = nn.BCELoss(reduction='sum')
    return loss(reconstruction.view(-1, 784), x.view(-1, 784))

def loss_vae(x, mu, logsigma, reconstruction):
    return KL_divergence(mu, logsigma) + log_likelihood(x, reconstruction)
```

И обучим модель:

```python
criterion = loss_vae

autoencoder = VAE().to(device)

optimizer = torch.optim.Adam(autoencoder.parameters())
```

```python
n_epochs = 25
train_losses = []
val_losses = []

for epoch in range(n_epochs):
    
    autoencoder.train()
    train_losses_per_epoch = []
    for batch, _ in train_loader:
        optimizer.zero_grad()
        mu, logsigma, reconstruction = autoencoder(batch.to(device))
        loss = criterion(batch.to(device).float(), mu, logsigma, reconstruction)
        loss.backward()
        optimizer.step()
        train_losses_per_epoch.append(loss.item())

        train_losses_per_epoch.append(loss.item())

    train_losses.append(np.mean(train_losses_per_epoch))

    autoencoder.eval()
    val_losses_per_epoch = []
    plt.figure(figsize=(16, 4))
    with torch.no_grad():
        for batch, _ in test_loader:
          mu, logsigma, reconstruction = autoencoder(batch.to(device))
          reconstruction = reconstruction.view(-1, 28, 28, 1)
          loss = criterion(batch.to(device).float(), mu, logsigma, reconstruction)
          val_losses_per_epoch.append(loss.item())
          for k in range(5):
            plt.subplot(2, 10, k+1)
            plt.imshow(batch[k][0])
            plt.title('Real')
            plt.axis('off')
            plt.subplot(2, 10, k+11)
            plt.title('Reconstruction')
            plt.axis('off')
            plt.imshow(reconstruction[k].cpu().detach().numpy())

    val_losses.append(np.mean(val_losses_per_epoch))  
    plt.suptitle('%d / %d - loss: %f' % (epoch+1, n_epochs, np.mean(val_losses_per_epoch)))
    plt.subplot(1, 2, 2)
    plt.plot(np.arange(len(train_losses)), train_losses, label='Train')
    plt.plot(np.arange(len(val_losses)), val_losses, label='Validation')
    plt.xlabel('Epoch')
    plt.title('loss')
    plt.legend()
    plt.show()    
    
    display.clear_output(wait=True)
```

### Архитектура

На картинке ниже представлена архитектура простого Conditional VAE.

По сути, единственное отличие от обычного -- это то, что мы вместе с картинкой в первом слое энкодера и декодера передаем еще информацию о классе картинки. 

То есть, в первый (входной) слой энкодера подается конкатенация картинки и информации о классе (например, вектора из девяти нулей и одной единицы). В первый слой декодера подается конкатенация латентного вектора и информации о классе.

На всякий случай: это VAE, то есть, latent у него все еще состоит из mu и sigma

Таким образом, при генерации новой рандомной картинки мы должны будем передать декодеру сконкатенированные латентный вектор и класс картинки.

P.S. Также можно передавать класс картинки не только в первый слой, но и в каждый слой сети. То есть на каждом слое конкатенировать выход из предыдущего слоя и информацию о классе.

```python
latent_dim = 4

class CVAE(nn.Module):
    def __init__(self):
        super(CVAE, self).__init__()

        # encoder
        self.conv1 = nn.Conv2d(in_channels=1, 
                               out_channels=64, 
                               kernel_size=4, 
                               stride=2, 
                               padding=1) 

        self.conv2 = nn.Conv2d(in_channels=64, 
                               out_channels=64*2, 
                               kernel_size=4, 
                               stride=2, 
                               padding=1) 

        self.fc_mu = nn.Linear(in_features=64*2*7*7+10, 
                               out_features=latent_dim)
        self.fc_logvar = nn.Linear(in_features=64*2*7*7+10, 
                                   out_features=latent_dim)
            
        self.activation = nn.ReLU()

        # decoder
        self.fc_d = nn.Linear(in_features=latent_dim+10, 
                            out_features=64*2*7*7)

        self.conv2_d = nn.ConvTranspose2d(in_channels=64*2, 
                                        out_channels=64, 
                                        kernel_size=4, 
                                        stride=2, 
                                        padding=1)
        self.conv1_d = nn.ConvTranspose2d(in_channels=64, 
                                        out_channels=1, 
                                        kernel_size=4, 
                                        stride=2, 
                                        padding=1)
        self.activation_d = nn.ReLU()
        
    def encode(self, x, n_class):
        x = self.activation(self.conv1(x))
        x = self.activation(self.conv2(x))
        x1 = x.view(x.shape[0], -1)
        x2 = n_class
        x = torch.cat((x1, x2), dim=1)
        x_mu = self.fc_mu(x)
        x_logvar = self.fc_logvar(x)
        return x_mu, x_logvar, n_class
     
    def decode(self, x, n_class):
        x = torch.cat((x, n_class), dim=1)
        x = self.fc_d(x)
        x = x.view(x.size(0), 64*2, 7, 7)
        x = self.activation_d(self.conv2_d(x))
        x = torch.sigmoid(self.conv1_d(x))
        return x

    def gaussian_sampler(self, mu, x_logvar):
        if self.training:
                std = (x_logvar * 0.5).exp()  
                return torch.distributions.Normal(loc=mu, scale=std).rsample()
        else:
                return mu

    def reparameterize(self, mu, x_logvar):
        std = torch.exp(0.5 * x_logvar) # standard deviation
        eps = torch.randn_like(std) # `randn_like` as we need the same size
        sample = mu + (eps * std) # sampling as if coming from the input space
        return sample

    def get_latent_vector(self, x, n_class):
        mu, x_logvar, n_class = self.encode(x, n_class)
        z = self.reparameterize(mu, x_logvar)
        return z

    def forward(self, x, n_class):
        mu, x_logvar, n_class = self.encode(x, n_class)
        reconstruction = self.decode(self.gaussian_sampler(mu, x_logvar), n_class)
        return mu, x_logvar, reconstruction
```
